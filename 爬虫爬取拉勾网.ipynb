```python
import requests
import jsonpath
import pandas as pd
kw=input('请输入工作:')
i=1
jobss=[]
while i <31:
    url='https://www.lagou.com/jobs/positionAjax.json?px=default&city=郑州&needAddtionalResult=false'
    data={'first': 'true',
          'pn': i,
          'kd': kw,
            }
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36',
             'Referer': 'https://www.lagou.com/jobs/list_python?labelWords=&fromSearch=true&suginput=',
             'Accept': 'application/json, text/javascript, */*; q=0.01',
             'Connection': 'keep-alive',
             
            }
    
    session=requests.session()
    session.get('https://www.lagou.com/jobs/list_python?labelWords=&fromSearch=true&suginput=',headers=headers)
    cookies=session.cookies
    #print(cookies.get_dict())
    res=requests.post(url=url,headers=headers,data=data,cookies=cookies)
    result=res.json()
    jobs=result['content']['positionResult']['result']
    
    jobss+=jobs
    i+=1
cities=jsonpath.jsonpath(jobss,'$..city')
#     print(cities)
districts=jsonpath.jsonpath(jobss,'$..district')
positionName=jsonpath.jsonpath(jobss,'$..positionName')
companyFullNames=jsonpath.jsonpath(jobss,'$..companyFullName')
industryField=jsonpath.jsonpath(jobss,'$..industryField')
companySize=jsonpath.jsonpath(jobss,'$..companySize')
workYear=jsonpath.jsonpath(jobss,'$..workYear')
education=jsonpath.jsonpath(jobss,'$..education')
positionAdvantage=jsonpath.jsonpath(jobss,'$..positionAdvantage')
skillLables=jsonpath.jsonpath(jobss,'$..skillLables')
financeStage=jsonpath.jsonpath(jobss,'$..financeStage')
salaries=jsonpath.jsonpath(jobss,'$..salary')
#print(financeStage)
# print(positionName)
# print(salaries)
data={'城市':cities,'区域':districts,'职位':positionName,'技能要求':skillLables,'薪资':salaries,'工作年限':workYear,'学历':education,'公司全称':companyFullNames,'行业':industryField,'规模':companySize,'福利待遇':positionAdvantage,'经济实力':financeStage}
datas=pd.DataFrame(data)
datas.to_excel('职工表0.xlsx',sheet_name='综合表')   #用pd 保存部分数据

# =============================================================================
#     datas=pd.DataFrame(jobs)  #以excel的格式保存在不同的工作薄中
#     datas.to_excel('{}.xlsx'.format(i))
# =============================================================================
# =============================================================================
#     data=str(jobs)
#     with open('测试.text','a') as f:  #以列表的方式保存为TXT格式
#         f.write(data)
# =============================================================================
      
# =============================================================================
#     cities=jsonpath.jsonpath(jobs,'$..city')
#     print(cities)
#     positionName=jsonpath.jsonpath(jobs,'$..positionName')
#     print(positionName)
#     salaries=jsonpath.jsonpath(jobs,'$..salary')
#     print(salaries)
#     with open('职位集合.xlsx','a+') as f:  # 以with open 的方式保存部分数据
#         f.write("城市:{},职位:{},薪资:{}".format(cities,positionName,salaries))
# =============================================================================




    


```
